Добавить новый столбец NEW, который вынести  значение столбца AAPL, если оно больше среднего арифметического столбца AAPL и значение равное 0, в ином случае.

```python
aapl_mean = dataframe["AAPL"].mean()
print("Среднее значение по столбцу AAPL = ", aapl_mean)
dataframe['NEW'] = np.where(
  dataframe['AAPL'] > aapl_mean, dataframe['AAPL'], 0
)
```
